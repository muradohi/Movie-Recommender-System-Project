```python

def get_genres(x):
    gen = []
    st = ' '
    for i in ast.literal_eval(x):
        gen.append(i.get('name'))
        
    if gen==[]:
        return np.NaN
    else:
        return (st.join(gen))
        
        
    
    
```

```python
data['genres'] = data['genres'].map(lambda x: get_genres(x))
```

```python
data['cast'][0]
```

```python
for i in ast.literal_eval(data['cast'][0]):
    names=[]
    names.append(i.get('name'))
    print(names[0])
```

```python
def get_actor1(x):
    actor1 = []
    for i in ast.literal_eval(x):
        actor1.append(i.get('name'))
        
    if actor1 == []:
        return np.NaN
    else:
        return(actor1[0])
    
def get_actor2(x):
    actor2 = []
    for i in ast.literal_eval(x):
        actor2.append(i.get('name'))
        
    if actor2 == [] or len(actor2)<=1:
        return np.NaN
    else:
        return(actor2[1])
    
def get_actor3(x):
    actor3 = []
    for i in ast.literal_eval(x):
        actor3.append(i.get('name'))
        
    if actor3 == [] or len(actor3)<=2:
        return np.NaN
    else:
        return(actor3[2])    
        
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: get_actor1(x))
data['actor_2_name'] = data['cast'].map(lambda x: get_actor2(x))
data['actor_3_name'] = data['cast'].map(lambda x: get_actor3(x))
```

```python
for i in ast.literal_eval(data['crew'][0]):
    if i.get('job') =='Director':
        print(i.get('name'))
```

```python
def get_director(x):
    director = []
    st = ' '
    for i in ast.literal_eval(x):
        if i.get('job')=='Director':
            director.append(i.get('name'))
        
    if director == []:
        return np.NaN
    else:
        return(st.join(director))
```

```python
data['director'] = data['crew'].map(lambda x: get_director(x))
```

```python
data['tags'] = data['title_x']+' '+data['genres']+' '+data['actor_1_name']+' '+data['actor_2_name']+' '+data['actor_3_name']+' '+data['director']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python


vector = cv.fit_transform(data['tags'].values.astype('U')).toarray()

```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vector)

```

```python


data[data['title'] == 'Toy Story'].index[0]

```

```python
def recommend(movie):
    index = data[data['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(data.iloc[i[0]].title)
```

```python
recommend('Toy Story')
```
